```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# pdf(probability density function
x=np.linspace(-4,4,200)
y=norm.pdf(x,0,1)
plt.plot(x,y)
plt.title("pdf(probability density function)",color="white")
plt.grid()
plt.xlabel("x",color="white")
plt.ylabel("y",color="white")
plt.show()
```

```python
# cdf (Cumulative Distribution Function)
x=np.linspace(-4,4,200)
y=norm.cdf(x,0,1)
plt.plot(x,y)
plt.title("cdf(Cumulative Distribution Function)",color="white")
plt.grid()
plt.xlabel("x",color="white")
plt.ylabel("y",color="white")
plt.show()
```
